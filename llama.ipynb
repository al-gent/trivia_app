```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load LLaMA model and tokenizer from Hugging Face
model_name = "meta-llama/Llama-7b-hf"  # Or any other version you want
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Define a prompt (in this case, summarizing why an article is trending)
prompt = "Summarize the reason why this Wikipedia article is currently trending:"

# Tokenize the input
inputs = tokenizer(prompt, return_tensors="pt")

# Generate a response
outputs = model.generate(inputs['input_ids'], max_length=150)

# Decode and print the response
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
